```python
plt.figure(figsize=(8,6))
points = np.random.poisson(lam=5, size=1000)
sns.distplot(points, norm_hist=True)
plt.show()
```

```python
samples = [5, 10, 50]

for sample in samples: 
    sample_list = [np.mean(np.random.poisson(lam=5, size=5)) for i in range(1000)]
    
    plt.figure(figsize=(8,6))
    sns.distplot(sample_list, norm_hist=True)
    plt.title("Sample size - {}".format(sample))
    plt.show()
    
    mean_sample = round(np.mean(sample_list), 4)
    median_sample = round(np.median(sample_list), 4)
    std_sample = round(np.std(sample_list), 4)
    
    print("Mean - {}".format(mean_sample))
    print("Median - {}".format(median_sample))
    print("STD - {}".format(std_sample))
```
