```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from scipy import linalg 

%matplotlib inline
```

```python
def func(x):
    return np.sin(x/5) * np.exp(x/10) + 5*np.exp((-x)/2)
```

<b>График функции f(x)</b>

```python
f_x = [func(x) for x in range(16)]

plt.figure(figsize=(8, 6))
plt.plot(range(16), f_x)
plt.show()
```

<b>График многочелна первой степени<b>

```python
points = [1,15]
A = []
for i in range(len(points)):
    A.append([points[i]**n for n in range(len(points))])
A = np.array(A)
b = [func(x) for x in points]

dots_1 = linalg.solve(A,b)

plt.figure(figsize=(8, 6))
plt.plot(points, dots_1, label="Многочлен первой степени")
plt.plot(range(16), f_x, label="f (x)")
plt.legend(loc='best')
plt.show()

print(dots_1)
```

<b>График многочлена второй степени</b>

```python
points = [1,8,15]
A = []
for i in range(len(points)):
    A.append([points[i]**n for n in range(len(points))])
A = np.array(A)
b = [func(x) for x in points]
dots_2 = linalg.solve(A,b)

plt.figure(figsize=(8,6))
plt.plot(points, dots_2, label="Многочлен второй степени")
plt.plot(range(16), f_x, label="f (x)")
plt.legend(loc="best")
plt.show()

print(dots_2)
```

<b>График многочлена третьей степени</b>

```python
points = [1,4,10,15]
A = []
for i in range(len(points)):
    A.append([points[i]**n for n in range(len(points))])
A = np.array(A)
b = [func(x) for x in points]
dots_3 = linalg.solve(A,b)

plt.figure(figsize=(8,6))
plt.plot(points, dots_3, label="Многочлен третьей степени")
plt.plot(range(16), f_x, label="f (x)")
plt.legend(loc="best")
plt.show()

print(dots_3)
```
