```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from scipy.optimize import minimize, differential_evolution
```

<h3>Минимизация негладкой функции</h3>

```python
def f_x(x):
    return int(np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2))
```

```python
y = [f_x(x) for x in range(1,31)]
x = list(range(1,31))

plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.title("Исходная функция")
plt.show()
```

```python
min_1 = minimize(f_x, 30, method="BFGS")
answer_1 = round(min_1.fun, 2)

print("Answer 1: {}".format(answer_1))
```

```python
points = [(1,30)]
min_2 = differential_evolution(f_x, points)
answer_2 = round(min_2.fun, 2)

print("Answer 2: {}".format(answer_2))
```

```python
answer_array = [answer_1, answer_2]

with open("week 3, task 3.txt", "w") as f:
    for answer in answer_array:
        f.write(str(answer) + " ") 
```
