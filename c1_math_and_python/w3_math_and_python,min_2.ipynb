```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from scipy.optimize import differential_evolution
```

<h3>Глобальная оптимизация</h3>

```python
def f_x(x):
    return (np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2))
```

```python
y = [f_x(x) for x in range(1,31)]
x = list(range(1,31))

plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.title("Исходная функция")
plt.show()
```

```python
points = [(1,30)]
minimize = differential_evolution(f_x, points)

answer = round(minimize.fun[0], 2)
print("Answer: {}".format(answer))
```

```python
with open("week 3, task 2.txt", "w") as f:
    f.write(str(answer))
```
