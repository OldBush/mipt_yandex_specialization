```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from scipy.optimize import minimize
```

<h3> Минимизация гладкой функции</h3>

```python
def f_x(x):
    return (np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2))
```

```python
y = [f_x(x) for x in range(1,31)]
x = list(range(1,31))

plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.title("Исходная функция")
plt.show()
```

```python
# Пример использования minimize

x = round(minimize(f_x, 7).x[0], 4)
y = round(minimize(f_x, 7).fun, 4)

print("Знаечние функции: {}; Значение аргумента: {}".format(y, x))
```

```python
minimize_1 = minimize(f_x, 2, method = "BFGS")
answer_1 = round(minimize_1.fun, 2)

print("Answer 1: {}".format(answer_1))
```

```python
minimize_2 = minimize(f_x, 30, method="BFGS")
answer_2 = round(minimize_2.fun, 2)

print("Answer 2: {}".format(answer_2, 2))
```

```python
answer_array = [answer_1, answer_2]

with open("week 3, task 1.txt", "w") as f:
    for answer in answer_array:
        f.write(str(answer) + " ")
```
